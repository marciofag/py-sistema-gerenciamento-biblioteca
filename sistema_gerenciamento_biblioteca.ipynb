```python
# Classe Livro
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

# Classe Usuario
class Usuario:
    def __init__(self, nome):
        self.nome = nome
        self.livros_emprestados = []

# Estruturas principais
livros = []
usuarios = []

# Fun√ß√£o: Cadastrar livro
def cadastrar_livro():
    titulo = input("T√≠tulo do livro: ")
    autor = input("Autor: ")
    genero = input("G√™nero: ")
    quantidade = int(input("Quantidade dispon√≠vel: "))
    livros.append(Livro(titulo, autor, genero, quantidade))
    print("‚úÖ Livro cadastrado com sucesso!\n")

# Fun√ß√£o: Listar livros
def listar_livros():
    if not livros:
        print("‚ö†Ô∏è Nenhum livro cadastrado.\n")
    else:
        for i, livro in enumerate(livros, 1):
            print(f"{i}. {livro.titulo} | {livro.autor} | {livro.genero} | Dispon√≠vel: {livro.quantidade}")
        print()

# Fun√ß√£o: Cadastrar usu√°rio
def cadastrar_usuario():
    nome = input("Nome do usu√°rio: ")
    usuarios.append(Usuario(nome))
    print("‚úÖ Usu√°rio cadastrado com sucesso!\n")

# Fun√ß√£o: Emprestar livro
def emprestar_livro():
    nome_usuario = input("Nome do usu√°rio: ")
    titulo_livro = input("T√≠tulo do livro: ")
    usuario = next((u for u in usuarios if u.nome == nome_usuario), None)
    livro = next((l for l in livros if l.titulo == titulo_livro), None)

    if usuario and livro and livro.quantidade > 0:
        livro.quantidade -= 1
        usuario.livros_emprestados.append(livro.titulo)
        print("‚úÖ Livro emprestado com sucesso!\n")
    else:
        print("‚ùå Usu√°rio ou livro n√£o encontrado, ou livro indispon√≠vel.\n")

# Fun√ß√£o: Devolver livro
def devolver_livro():
    nome_usuario = input("Nome do usu√°rio: ")
    titulo_livro = input("T√≠tulo do livro: ")
    usuario = next((u for u in usuarios if u.nome == nome_usuario), None)
    livro = next((l for l in livros if l.titulo == titulo_livro), None)

    if usuario and livro and titulo_livro in usuario.livros_emprestados:
        livro.quantidade += 1
        usuario.livros_emprestados.remove(titulo_livro)
        print("‚úÖ Livro devolvido com sucesso!\n")
    else:
        print("‚ùå Devolu√ß√£o inv√°lida.\n")

# Menu interativo
def menu():
    while True:
        print("\nüìö SISTEMA DE BIBLIOTECA")
        print("1. Cadastrar Livro")
        print("2. Listar Livros")
        print("3. Cadastrar Usu√°rio")
        print("4. Emprestar Livro")
        print("5. Devolver Livro")
        print("6. Sair")

        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            cadastrar_livro()
        elif opcao == "2":
            listar_livros()
        elif opcao == "3":
            cadastrar_usuario()
        elif opcao == "4":
            emprestar_livro()
        elif opcao == "5":
            devolver_livro()
        elif opcao == "6":
            print("üëã Encerrando o sistema. At√© logo!")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida. Tente novamente.")

# Executar menu
menu()
```
